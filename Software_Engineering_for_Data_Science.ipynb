Directory should look like this:  image.png



Package name should be short, all lower case, and only contains underscore: my_package


Minimum package requirements: Directory + init.py file


__init__.py: Required in the directory so that python will treat the directory as a package/ module that is importable.
(for the python versions < 3.3)

utils.py: A sub module existing in  root directory/package directory/utils.py. Defining the classes and methods here.

Example:

a) Referencing the utils module in my_script file

*In utils.py*

```
def a_method(a,b)
  if a>b:
    print('a is bigger than b')
```
*In my_script.py file*

```
import my_package.utils

# Checking if a > b

my_package.utils.a_method(a,b)

```

b) We can refer to utils module in __init__.py file and we will not need to refer to utils anymore:


*In init.py file*

`from .utils import a_method ` # Notice we used '.' before utils for the relative import. 

*In my_script.py file*

```
import my_package
my_package.a_method(a,b) #calling the method directly from the package.

```
We should import the key functionalities of the submodules in init.py.

6) Sharing the package:

Two main steps to sharing a python package: creating setup.py and requirements.txt.

These two pieces provide information on how to install your package and recreate its required environment. These files list information about what dependencies you've used as well as allowing you to describe your package with additional metadata.

Setup.py and requirements.txt should be located in the same directory as package. 


a) Example requirements.txt:



```
# Needed packages:

pandas
numpy == 1.15.4
pycodestyle >= 2.4.0
```

If we execute install in the terminal, it will create the necessary environment for the package:

```
pip install -r requirements.txt
```

b) Example creation of setup.py:

```
from setuptools import setup

setup(name='my_package' #directory name
      version = '0.0.2',
      description='Sevdos first package',
      author='Sevde Akkoc'
      author_email='sevdeakkoc@gmail.com',
      packages=['my_package'] #the location
      install_requires=['matplotlib', 'numpy==1.15.4', 'pycodestyle>=2.4.0']

```
